```python
import pandas as pd
import datetime
from datetime import date
Data=pd.read_csv('/content/drive/MyDrive/python/project python.csv')
Data
```

```python
data1=Data.groupby(['customer ID', 'Date']).agg(max=('Brand count', 'max'),total=('Brand count', 'sum')).reset_index()
data1
```

```python
data1['Percentage cutoff']= (data1['max']/data1['total'])*100
data1
```
